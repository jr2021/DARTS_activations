```python
import matplotlib.pyplot as plt
import json
```

```python
name = 'run/cifar10/bananas/0/errors.json'

with open(name) as f:
    errors = json.load(f)
```

```python
plt.plot(errors['train_loss'], label='train')
plt.plot(errors['valid_loss'], label='valid')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.legend()
plt.show()
```

```python
plt.plot(errors['train_acc'], label='train')
plt.plot(errors['valid_acc'], label='valid')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
